```python
import pickle
import os
from sklearn.svm import SVC
import cv2
import numpy as np
from sklearn.metrics import classification_report
```

```python
# load the model from disk
filename = './models/model_face_not.sav'
clf = pickle.load(open(filename, 'rb'))
# load pca
filename = './models/pca_face_not.sav'
pca = pickle.load(open(filename, 'rb'))

def normailze_data_face(data):
    data = data.astype('float64')
    data /= 255
    return data
```

```python
# video = cv2.VideoCapture('./data_true/IMG_5107.MOV')
video = cv2.VideoCapture(0)
face_classifier = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

def draw_bounding_box(vid, bounding_box,label, y_pred):
    
    for i in range(len(bounding_box)):
        x = bounding_box[i][0]
        y = bounding_box[i][1]
        w = bounding_box[i][2]
        h = bounding_box[i][3]
        # print(x)
        cv2.rectangle(vid, (x, y), (x + w, y + h), (0, 255, 0), 4)
        if y_pred[i] == 1:
            cv2.putText(vid, label, (x, y - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
        else:
            cv2.putText(vid, 'Unknown', (x, y - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

def detect_bounding_box(vid):
    # scaleFactor giam kich thuoc cua anh down 10%
    faces = face_classifier.detectMultiScale(vid, scaleFactor=1.1, minNeighbors=9, minSize=(64, 64))
    # cut face
    if len(faces) == 0:
        return 0, 0
    # return list face
    img_faces = []
    for (x, y, w, h) in faces:
        img_face = vid[y:y + h, x:x + w]
        img_face = cv2.resize(img_face, (64, 64))
        img_faces.append(img_face)
    
    return np.array(img_faces), faces

check = 1
while True:
    ret, frame = video.read()
    if ret:
        # frame = cv2.resize(frame, (640, 480))
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        # draw bounding box and return face cutting
        faces, coor_faces = detect_bounding_box(frame)
        # print(faces.shape)
        
        if type(faces) == int:
            cv2.imshow("My Face Detection Project", frame) 
            continue
        
        # faces = faces.reshape(1, -1)
        faces = faces.reshape((faces.shape[0],faces.shape[1]*faces.shape[2]))
        faces = normailze_data_face(faces)
        faces = pca.transform(faces)
        y_pred = clf.predict(faces)
        
        
        draw_bounding_box(frame, coor_faces, 'Minh Anh', y_pred)
        print(y_pred)
        cv2.imshow(
            "My Face Detection Project", frame
        ) 
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
```
